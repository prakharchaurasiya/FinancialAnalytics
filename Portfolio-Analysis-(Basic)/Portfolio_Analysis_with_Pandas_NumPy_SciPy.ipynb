```python
import pandas as pd
import numpy as np

!wget -O 'Data for Portfolio Analysis.xlsx' https://drive.google.com/uc?id=1FbRYF6Gu-hisilRUjhh56Fj8AzE50h-P&export=download

data = pd.read_excel('Data for Portfolio Analysis.xlsx')
data['Date'] = pd.to_datetime(data['Date ']) 
data.index = data['Date']
data.drop(['Date ', 'Date'], axis=1,inplace=True)
```

```python
data
```

```python
daily_returns = data.pct_change().dropna()
daily_returns
```

```python
daily_log_returns = np.log(data) - np.log(data.shift(1))
daily_log_returns = daily_log_returns.dropna()
daily_log_returns
```

```python
average_returns = daily_log_returns.mean()
average_returns
```

```python
annual_returns = daily_log_returns.sum() # adding daily log returns gives annual return
annual_returns
```

```python
variance = daily_log_returns.var(ddof=0) # ddof=0 for population, ddof = 1 for sample
variance
```

```python
annual_variance = variance * 250
annual_variance
```

```python
standard_deviation = daily_log_returns.std(ddof=0) # ddof=0 for population, ddof = 1 for sample
standard_deviation
```

```python
annual_standard_deviation = standard_deviation*np.sqrt(250)
annual_standard_deviation
```

```python
covariance_matrix = daily_log_returns.cov(ddof=0) # ddof=0 for population, ddof = 1 for sample
covariance_matrix
```

```python
annual_covariance_matrix = covariance_matrix*250
annual_covariance_matrix
```

```python
PORTFOLIO = pd.DataFrame()
PORTFOLIO['Returns'] = annual_returns
equal_weights = [1/annual_returns.shape[0] for i in range(annual_returns.shape[0])]
PORTFOLIO['Weights'] = pd.Series(equal_weights, index=annual_returns.index)
PORTFOLIO 
```

```python
portfolio_return = np.dot((PORTFOLIO['Weights'].values).T, PORTFOLIO['Returns'].values)
portfolio_return
```

```python
portfolio_standard_deviation = np.sqrt(np.dot((PORTFOLIO['Weights'].values).T, np.dot(annual_covariance_matrix.values, PORTFOLIO['Weights'].values)))
portfolio_standard_deviation
```

```python
import scipy
from scipy.optimize import minimize, Bounds

# objective function
def sharpe_ratio(weights, covar_matrix, annual_returns, Rf=0.0633, sign=-1.0):
  std = np.sqrt(np.dot(np.dot(weights.T, covar_matrix), weights))
  mean = np.dot(weights.T, annual_returns)
  return sign*(mean - Rf)/std # multiplied by -1 because scipy don't have maximize function

def constraint(weights):
  return weights.sum() - 1

w0 = np.array([0.20, 0.30, 0.40, 0.10]) # random guess

cons = [{'type': 'eq', 'fun': constraint}]
bnds = Bounds(np.zeros(4),np.ones(4))

sol = minimize(sharpe_ratio, w0, args=(annual_covariance_matrix.values, annual_returns.values), method='SLSQP', constraints=cons, bounds=bnds)
sol
```

```python
np.round_(sol.x, decimals=4)
```

```python
PORTFOLIO['Optimal Weights'] = pd.Series(np.round_(sol.x, decimals=4), index=annual_returns.index)
PORTFOLIO[['Returns', 'Optimal Weights']]
```

```python
# Solution form excel
wi = np.array([0.1008, 0, 0.8992, 0])
rtn = np.array([0.2151, 0.1692, 0.3917, 0.0199])
sharpe_ratio(wi, annual_covariance_matrix.values, rtn, sign=1.0)
```
